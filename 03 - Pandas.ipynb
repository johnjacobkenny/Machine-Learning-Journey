```python
temperature_df = pd.DataFrame({
    'city': ['Trivandrum', 'Ernakulam', 'Thrissur'],
    'temperature': [30, 35, 38]
})
```

```python
humidity_df = pd.DataFrame({
    'city': ['Trivandrum', 'Ernakulam'],
    'humidity': [80, 90]
})
```

```python
pd.merge(temperature_df, humidity_df, on='city', how='outer')
```
