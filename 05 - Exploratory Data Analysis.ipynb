```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
iris = pd.read_csv("data/iris.csv")
```

```python
iris.columns
```

```python
iris["species"].value_counts()

# balanced data set, all classes have a similar # of datapoints
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", size = 4)\
  .map(plt.scatter, "sepal_length", "sepal_width")\
  .add_legend()
    
plt.show()
# linearly separable
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", size=5)\
 .map(plt.scatter, "petal_length", "petal_width")\
 .add_legend()

plt.show()
```

```python
# pair plot which plots 2 variables
sns.pairplot(iris, hue="species", size=3)
plt.show()
```

This only gets us to some place. For data sets which have more dimensions we will need other ways to reduce the dimensionality.

```python
sns.FacetGrid(iris, hue="species", size=5)\
 .map(sns.distplot, "petal_length")\
 .add_legend()
plt.show()

# Probability density function (PDF) is a smoothed histogram using KDE (kernel density estimation)
```

```python
sns.FacetGrid(iris, hue="species", size=10)\
 .map(sns.distplot, "petal_width")\
 .add_legend()
plt.show()
```
